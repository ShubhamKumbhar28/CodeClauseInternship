```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("/content/heart_disease_data.csv")
```

```python
data=data.drop_duplicates()
```

```python
data.describe()
```

```python
# check distribution of target variable
data['target'].value_counts().plot(kind='bar')
```

```python
# Split the features into X and y
X=data.drop(columns='target',axis=1)
```

```python
y=data['target']
```

```python
X.head()
```

```python
y.head()
```

```python
 #split data into training and testing

X_train,X_test ,y_train, y_test = train_test_split(X,y,random_state=10,test_size=0.3,stratify=y)
```

```python
#Model training
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
# Predict on test data
y_pred=model.predict(X_test)

```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#predicting

data=(52,1,2,172,199,1,1,162,0,0.5,2,0,3)

inp_data=np.asarray(data)
re_data=inp_data.reshape(1,-1)

prediction=model.predict(re_data)
```

```python
print(prediction)

if prediction[0]==0:
  print('No heart diseae')
else:
  print('heart disease')
```
